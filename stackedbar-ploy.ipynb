```python
months = []
for i in range(1,13):
    day = date(2024,i,1).strftime('%B')
    months.append(day)
```

```python
# months
```

```python
sales = np.random.randint(100,200,12)
```

```python
len(sales)
```

```python
plt.figure(figsize=(12,4))
plt.bar(x=months,height=sales,alpha=0.2,edgecolor='black')
plt.plot(months,sales,color='red',marker='o')

for i in range(0,len(months)):
    plt.annotate(xy=(i,sales[i]/2),
                 text=months[i],
                 horizontalalignment='center')
    
    plt.annotate(xy=(i,sales[i]),
                 text=sales[i],
                 horizontalalignment='center')

plt.xlabel('Months')
plt.xticks(rotation=30)

plt.show()
```

```python
plt.figure(figsize=(12,4))
cm = plt.cm.plasma(np.linspace(0.2,.8,12))

plt.bar(x=months,height=sales,alpha=0.2,edgecolor='black',color=cm)
plt.plot(months,sales,color='red',marker='o')

for i in range(0,len(months)):
    plt.annotate(xy=(i,sales[i]/2),
                 text=months[i],
                 horizontalalignment='center')
    
    plt.annotate(xy=(i,sales[i]),
                 text=sales[i],
                 horizontalalignment='center')

plt.xlabel('Months')
plt.xticks(rotation=30)

plt.show()
```
